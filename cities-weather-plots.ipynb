## Starter Code to Generate Random Geographic Coordinates and a List of Cities

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
from pprint import pprint
from scipy import stats
from colorspacious import cspace_converter
```

### Read in CSV file

```python
# Read csv file into pandas dataframe
city_data = pd.read_csv ('Resources/cities.csv')

# Display sample data
city_data.head()
```

## Requirement 1: Create Plots to Showcase the Relationship Between Weather Variables and Latitude

#### Latitude Vs. Temperature

```python
# Build scatter plot for latitude vs. temperature

x = city_data["Lat"]
y = city_data["Max Temp"]

plt.scatter(x, y, c=y, cmap="turbo")

# Incorporate the other graph properties

plt.title("Latitude vs Max Temperature of Cities")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.colorbar()
plt.grid()

# Save the figure
plt.savefig("assets/images/lat_vs_temp.png")

# Show plot
plt.show()
```

#### Latitude Vs. Humidity

```python
# Build the scatter plots for latitude vs. humidity

x = city_data["Lat"]
y = city_data["Humidity"]

plt.scatter(x, y, c=y, cmap="turbo")

# Incorporate the other graph properties

plt.title("Latitude vs Humidity of Cities")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.colorbar()
plt.grid()

# Save the figure
plt.savefig("assets/images/lat_vs_humidity.png")

# Show plot
plt.show()
```

```python
# Build the scatter plots for latitude vs. cloudiness

x = city_data["Lat"]
y = city_data["Cloudiness"]

plt.scatter(x, y, c=y, cmap="turbo")

# Incorporate the other graph properties

plt.title("Latitude vs Cloudiness of Cities")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.colorbar()
plt.grid()

# Save the figure
plt.savefig("assets/images/lat_vs_cloudiness.png")

# Show plot
plt.show()
```
